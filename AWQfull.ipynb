```python
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer
from peft import PeftModel

# 1. Load base LLaMA 3.1 model
base_model_id = "meta-llama/Llama-3.1-8B-Instruct"
lora_adapter_id = "Richard9905/Lora_Bible_adaptors_forLLAMA3"
merged_model_path = "Merged_base_LLAMA3_bible"

# 2. Load base model in float16 (GPU assumed)
base_model = AutoModelForCausalLM.from_pretrained(
    base_model_id,
    torch_dtype=torch.float16,
    device_map="auto"
)

# 3. Load LoRA adapters
model = PeftModel.from_pretrained(base_model, lora_adapter_id)

# 4. Merge the adapters into the base weights
model = model.merge_and_unload()

# 5. Save the merged model
model.save_pretrained(merged_model_path, safe_serialization=True)

# 6. Save tokenizer too (optional but recommended)
tokenizer = AutoTokenizer.from_pretrained(base_model_id)
tokenizer.save_pretrained(merged_model_path)

print(f"✅ Merged model saved to: {merged_model_path}")
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM
from auto_gptq import AutoGPTQForCausalLM, BaseQuantizeConfig

merged_model_path = "Richard9905/Merged_base_LLAMA3_bible"  # <-- Define this!

# Load tokenizer and base (non-quantized) model
tokenizer = AutoTokenizer.from_pretrained(merged_model_path)
model = AutoModelForCausalLM.from_pretrained(merged_model_path)

# Define quantization config (4-bit example)
quantize_config = BaseQuantizeConfig(
    bits=4,              # 4-bit quantization
    group_size=128,      # typical group size for LLaMA
    desc_act=False,      # usually False
    sym=True             # symmetric quantization
)

# Load the model with quantization config for GPTQ
model = AutoGPTQForCausalLM.from_pretrained(
    merged_model_path,
    quantize_config=quantize_config,
    trust_remote_code=True
)

# Optional calibration texts for better quantization quality
calibration_texts = [
    "In the beginning God created the heaven and the earth.",
    "Jesus said, I am the way, the truth, and the life.",
    "Explain to me the story of Revelation from the beginning of Genesis to the end of Revelation."
]

tokenized_calibration = [tokenizer(text, return_tensors="pt") for text in calibration_texts]

# Perform quantization
model.quantize(tokenized_calibration)

# Save quantized model and tokenizer
quantized_model_path = "./Merged_base_LLAMA3_bible_4bit"
model.save_quantized(quantized_model_path)
tokenizer.save_pretrained(quantized_model_path)

print("✅ Quantized model saved at:", quantized_model_path)
```

```python
ffrom transformers import AutoTokenizer
from auto_gptq import AutoGPTQForCausalLM
import torch

# Load model and tokenizer
model_id = "Richard9905/4_bit_AWQ_Model"

tokenizer = AutoTokenizer.from_pretrained(model_id)
model = AutoGPTQForCausalLM.from_quantized(model_id, trust_remote_code=True)

# Bible context dictionary
bible_context = {
    "John 8:58": (
        "In John 8:58, Jesus says, 'Before Abraham was, I AM.' "
        "This statement uses the divine name 'I AM', linking Jesus to God's identity in Exodus 3:14."
    ),
    "Exodus 3:14": (
        "In Exodus 3:14, God tells Moses, 'I AM WHO I AM.' "
        "This phrase reveals God's eternal and self-existent nature. Jesus references this in John 8:58."
    ),
    "Genesis 1:1": (
        "In Genesis 1:1, the Bible begins with 'In the beginning God created the heavens and the earth,' "
        "affirming God's role as the Creator of everything."
    ),
    "Psalm 23:1": (
        "'The Lord is my shepherd; I shall not want' (Psalm 23:1) conveys God's guidance, provision, and care."
    ),
    "John 3:16": (
        "John 3:16 expresses God's love: 'For God so loved the world, that he gave his only Son...' "
        "It highlights the offer of eternal life through faith."
    ),
    # Add more verses as needed
}

# Function to answer user queries with limited context
def contextual_biblical_agent(user_query):
    device = "cuda" if torch.cuda.is_available() else "cpu"
    inputs = tokenizer(user_query, return_tensors="pt").to(device)
    model.to(device)
    outputs = model.generate(**inputs, max_new_tokens=512)
    base_response = tokenizer.decode(outputs[0], skip_special_tokens=True)

    # Collect up to 3 relevant verses only
    matched_verses = []
    for verse in bible_context:
        if len(matched_verses) >= 3:
            break
        if verse in base_response or verse.lower() in user_query.lower():
            matched_verses.append(verse)

    if matched_verses:
        contextual_notes = "\n\n".join(
            f"📖 **Contextual Explanation for {verse}**:\n{bible_context[verse]}"
            for verse in matched_verses
        )
        return f"{base_response}\n\n{contextual_notes}"
    else:
        return base_response

# Run the chatbot interactively
if __name__ == "__main__":
    while True:
        user_query = input("Ask your Bible question (or type 'exit' to quit): ")
        if user_query.lower() == "exit":
            break
        response = contextual_biblical_agent(user_query)
        print("\n" + response + "\n" + "-"*80 + "\n")
```

```python
import gradio as gr
from transformers import AutoTokenizer
from auto_gptq import AutoGPTQForCausalLM
import torch

# Load model and tokenizer
model_id = "Richard9905/4_bit_AWQ_Model"
tokenizer = AutoTokenizer.from_pretrained(model_id)
model = AutoGPTQForCausalLM.from_quantized(model_id, trust_remote_code=True)

# Optional: Set pad_token for smoother batch input if needed
tokenizer.pad_token = tokenizer.eos_token

# Bible context
bible_context = {
    "John 8:58": (
        "In John 8:58, Jesus says, 'Before Abraham was, I AM.' "
        "This statement uses the divine name 'I AM', linking Jesus to God's identity in Exodus 3:14."
    ),
    "Exodus 3:14": (
        "In Exodus 3:14, God tells Moses, 'I AM WHO I AM.' "
        "This phrase reveals God's eternal and self-existent nature. Jesus references this in John 8:58."
    ),
    "Genesis 1:1": (
        "In Genesis 1:1, the Bible begins with 'In the beginning God created the heavens and the earth,' "
        "affirming God's role as the Creator of everything."
    ),
    "Psalm 23:1": (
        "'The Lord is my shepherd; I shall not want' (Psalm 23:1) conveys God's guidance, provision, and care."
    ),
    "John 3:16": (
        "John 3:16 expresses God's love: 'For God so loved the world, that he gave his only Son...' "
        "It highlights the offer of eternal life through faith."
    ),
    # Add more as needed
}

# Contextual Bible QA logic
def contextual_biblical_agent(user_query):
    device = "cuda" if torch.cuda.is_available() else "cpu"
    inputs = tokenizer(user_query, return_tensors="pt").to(device)
    model.to(device)
    outputs = model.generate(**inputs, max_new_tokens=512)
    base_response = tokenizer.decode(outputs[0], skip_special_tokens=True)

    matched_verses = []
    for verse in bible_context:
        if len(matched_verses) >= 3:
            break
        if verse in base_response or verse.lower() in user_query.lower():
            matched_verses.append(verse)

    if matched_verses:
        contextual_notes = "\n\n".join(
            f"📖 **Contextual Explanation for {verse}**:\n{bible_context[verse]}"
            for verse in matched_verses
        )
        return f"{base_response}\n\n{contextual_notes}"
    else:
        return base_response

# Gradio interface
demo = gr.Interface(
    fn=contextual_biblical_agent,
    inputs=gr.Textbox(lines=2, placeholder="Ask a Bible question..."),
    outputs=gr.Textbox(lines=15),
    title="📖 Contextual Bible Chatbot",
    description="Ask Bible-based questions with contextual verse explanations (up to 3 references)."
)

if __name__ == "__main__":
    demo.launch()
```
